### <u>List slicing</u> 
**Similar to String slicing**      
   
       
list_name[ starting_idx : ending_idx]  #ending idx is not included         
marks = [87,64,33,95,76]     
marks[1:4] is [64,33,95]    
marks[:4] is same as marks[0:4]     
marks[1:] is same as marks[1:len(marks)]    
marks[-3:-1] is [33,95] 

```python
marks = [87,64,33,95,76]     
print(marks[1:4])    
print(marks[:4])  
print(marks[0:4])     
print(marks[1:]) 
print(marks[1:len(marks)])    
print(marks[-3:-1]) 
```

### <u> List Methods </u> 

list = [2,1,3]      

list.append(4)   #adds one element at the end *[2,1,3,4]*    

list.sort()      #sort in aseending order *[1,2,3]*    

list.sort(reverse = True)  #sorts in desending order *[3,2,1]*     

list.reverse()   #reverses list  *[3,1,2]    

list.insert(idx,el)  #insert element at index      

list = [2,1,3,1]     

list.remove(1)   #removes first occurrence of element [2,3,1]    

list.pop(idx)    #remove element at idx     



```python
list = [2,1,3]      

list.append(4)   #adds one element at the end *[2,1,3,4]*    
print(list)
list.sort()      
```

```python
list = [2,1,3]  
print(list.sort())      #sort in aseending order *[1,2,3]*    
print(list)
```

```python
list = [2,1,3]  
list.append(4)   
print(list.sort())         
print(list)
```

```python
list = [2,1,3]  
list.append(4)
list.sort(reverse = True)  #sorts in desending order [3,2,1]*    
print(list)
```

```python
list = [2,1,3]  
list.reverse()   #reverses list  [3,1,2]    
print(list)
```

```python
list = [2,1,3]  
list.insert(1,5)  #inserts element 5 at index 1 *[2,5,1,3,]*
print(list)
```

```python
list = [2,1,3,1]  
list.remove(1)   #removes first occurrence of element 1 *[2,3,1]*
print(list)
```

```python
list = [2,1,3,1]  
list.pop(2) 
print(list)  
```

### <u> Tuples in python </u>    
*A built-in data type that lets us cresat immutable sequences of values.*   

tup = (87,64,33,95,76)  #tup[0],tup[1]

tup[0] = 43  # NOT allowed in python   


tup1= ()

tup3=(1,2,3)  



```python
tup = (1,2,3,4,5)
print(tup)
print(type(tup))
print(tup[0])
print(tup[1]) 
print(tup[2])
```

```python
tup = () 
print(tup)
print(type(tup))
```

```python
tup = (1,)
print(tup)
print(type(tup))
```

```python
tup = (1)
print(tup)
print(type(tup))
```

```python
tuple = (2,1,3,4,5)
print(tuple) 
print(tuple[1:3])
```

### <u>Tuple Methods</u>

tup = (2,1,3,1)  

tup.index(el)  #returns index of first occurrence      

tup.count(el)  #count total occurrences            tup.count(1) is 2

```python
tuple = (2,1,3)
print(tuple.index(1))  #returns index of first occurrence of element 1 *[1]*
```

```python
tuple = (2,1,3)
print(tuple.index(2))
```

```python
tuple = (2,1,3)
print(tuple.count(1))
```

```python
tuple = (2,1,3,3)
print(tuple.count(3))
```
